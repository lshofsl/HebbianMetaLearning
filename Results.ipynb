```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# load fitness history
fitness_cartpole_ABCD_lr = np.load("/home/ubuntu/PMLS/HebbianMetaLearning/heb_coeffs/1769284315/Fitness_values_1769284315_CartPole-v0.npy")
fitness_cartpole_ABCD = np.load("/home/ubuntu/PMLS/HebbianMetaLearning/heb_coeffs/1769285901/Fitness_values_1769285901_CartPole-v0.npy")

std_cartpole_ABCD = np.load("/home/ubuntu/PMLS/HebbianMetaLearning/heb_coeffs/1769285901/Fitness_std_values_1769285901_CartPole-v0.npy")
std_cartpole = np.load( "/home/ubuntu/PMLS/HebbianMetaLearning/heb_coeffs/1769774300/Fitness_std_values_1769774300_CartPole-v0.npy")

plt.figure(figsize=(6,4))
plt.plot(fitness_cartpole_ABCD_lr, lw=2)
plt.fill_between(range(len(std_cartpole)), fitness_cartpole_ABCD_lr - std_cartpole, fitness_cartpole_ABCD_lr + std_cartpole, color='blue', alpha=0.3)
plt.plot(fitness_cartpole_ABCD, lw=2)
plt.fill_between(range(len(std_cartpole_ABCD)), fitness_cartpole_ABCD - std_cartpole_ABCD, fitness_cartpole_ABCD + std_cartpole_ABCD, color='orange', alpha=0.3)
plt.xlabel("Generation")
plt.ylabel("Mean reward")
plt.title("Evolution of fitness CartPole-v0")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Hebb_coeff_cartpole_ABCD_lr = np.load("heb_coeffs/1769284315/Hebbian_coefficients_history_1769284315_CartPole-v0.npy")
mean_hebb_coeff_cartpole_ABCD_lr = np.mean(Hebb_coeff_cartpole_ABCD_lr, axis=1)

labels = ['A', 'B', 'C', 'D', 'lr']

plt.figure(figsize=(6,4))
for i in range(len(labels)): 
    plt.plot(mean_hebb_coeff_cartpole_ABCD_lr[:, i], label=labels[i])
plt.xlabel("Generation")
plt.ylabel("Mean Hebbian coefficient")
plt.title("Evolution of mean Hebbian coefficient CartPole-v0")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
